**Confounding variable** : A confounding variable is a factor that influences both input features(independent variables) and the target variable(dependent variable) in the model.
There are potential cases of confounding variables in this study, which include:
1. Distance from earth:
  Stars from the same constellations can be at very different distances. Including this feature might confuse the model, and it could learn patterns based on distance rather
  than true sky position.
2. Apparent brightness:
  Nearby stars tend to be brighter, so brightness may correlate with distance. The model could associate brightness with constellation membership incorrectly.

  Handling Confounders:
  Distance can be scaled carefully so as to ensure that RA and Dec would be the main predictors, and Apparent magnitude should be checked for correlation with RA/Dec and constellation labels.

**Detailed Data cleaning plan**


---

Before analyzing the constellation dataset, it is essential to ensure the data is clean, consistent, and ready for modeling. The following steps outline the planned cleaning process:

1. Rename Columns for Clarity :
Columns will be renamed to clearly indicate their meaning and units. For example, “Right Ascension” will be renamed to “RA_deg” (degrees), and “Distance” to “Distance_ly” (light-years). This improves readability and reduces ambiguity.

2. Convert Data Types :
Many columns currently store numeric data as text. These will be converted to appropriate numeric types, and coordinates will be converted from sexagesimal to decimal degrees to facilitate analysis.

3. Handle Missing Data :
Missing values are common in astronomical data. Numeric missing values will be imputed using the median to maintain realistic distributions. Categorical missing values will be replaced with “Unknown” to retain as much data as possible. Rows missing both coordinates will be removed, as position is critical for constellation prediction.

4. Remove Duplicates :
Duplicate rows will be identified and removed to ensure each star is uniquely represented, preventing bias in analysis.

5. Investigate and Handle Outliers :
Outliers will be detected using the Z-score method. Extreme values will be reviewed to distinguish between valid rare astronomical phenomena and data errors. Erroneous outliers will be corrected or removed, while valid ones will be flagged for special consideration.

6. Standardize Categorical Values :
Categorical columns will be standardized in spelling and formatting to ensure consistency and facilitate analysis.

7. Add Flags for Imputed or Missing Data :
Binary flags will be added to indicate imputed or missing values, aiding transparency and model interpretability.




```python
import pandas as pd
import numpy as np
import re

# Load the dataset
data = pd.read_csv('constellations.csv')


# Rename columns for clarity
data = data.rename(columns={
    'Right Ascension': 'RA_hms',
    'Declination': 'Dec_dms',
    'Visual Magnitude': 'Vmag',
    'Absolute Magnitude': 'AbsMag',
    'Distance': 'Distance_ly',
    'Spectral class': 'Spectral_Class',
    'Bayer': 'Bayer_Designation',
    'Flamsteed': 'Flamsteed_Number',
    'Variable Star': 'Variable_Type',
    'Henry Draper Number': 'HD_Number',
    'Hipparcos Number': 'Hip_Number'
})

# Robust conversion for RA/Dec
def hms_to_deg(hms_str):
    try:
        if pd.isna(hms_str):
            return np.nan
        # Accepts "00h 08m 23.17s", "00:08:23.17", "00 08 23.17"
        hms_str = str(hms_str).replace('h', ' ').replace('m', ' ').replace('s', ' ')
        parts = re.split('[ :]+', hms_str.strip())
        parts = [p for p in parts if p]
        # Only proceed if first part is a number
        if not parts or not parts[0].replace('.', '', 1).isdigit():
            return np.nan
        h = float(parts[0])
        m = float(parts[1]) if len(parts) > 1 and parts[1].replace('.', '', 1).isdigit() else 0
        s = float(parts[2]) if len(parts) > 2 and parts[2].replace('.', '', 1).isdigit() else 0
        return h * 15 + m * 0.25 + s * 0.0041667
    except Exception:
        return np.nan

def dms_to_deg(dms_str):
    try:
        if pd.isna(dms_str):
            return np.nan
        # Accepts "+29° 05′ 27.0′′", "+29:05:27.0", "+29 05 27.0"
        dms_str = str(dms_str).replace('−', '-').replace('°', ' ').replace('′', ' ').replace('\'', ' ').replace('\"', ' ').replace('″', ' ')
        parts = re.split('[ :]+', dms_str.strip())
        parts = [p for p in parts if p]
        if not parts or not parts[0].replace('.', '', 1).replace('-', '', 1).isdigit():
            return np.nan
        sign = -1 if '-' in parts[0] else 1
        d = abs(float(parts[0]))
        m = float(parts[1]) if len(parts) > 1 and parts[1].replace('.', '', 1).isdigit() else 0
        s = float(parts[2]) if len(parts) > 2 and parts[2].replace('.', '', 1).isdigit() else 0
        return sign * (d + m/60 + s/3600)
    except Exception:
        return np.nan

data['RA_deg'] = data['RA_hms'].apply(hms_to_deg)
data['Dec_deg'] = data['Dec_dms'].apply(dms_to_deg)


# Convert numeric columns
for col in ['Distance_ly', 'Vmag', 'AbsMag']:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Impute missing values with median (for numeric columns)
for col in ['RA_deg', 'Dec_deg', 'Distance_ly', 'Vmag', 'AbsMag']:
    data[col] = data[col].fillna(data[col].median())


# Only drop rows if both RA and Dec are missing (should be none after imputation)
data = data.dropna(subset=['RA_deg', 'Dec_deg'], how='all')


# Remove duplicates
data = data.drop_duplicates()


# Remove only extreme outliers (very high thresholds)
data = data[(data['AbsMag'] <= 1000) & (data['Distance_ly'] <= 1e8)]


# Fill categorical columns with 'Unknown'
categorical_cols = ['Spectral_Class', 'Bayer_Designation', 'Flamsteed_Number', 'Variable_Type', 'HD_Number', 'Hip_Number']
for col in categorical_cols:
    data[col] = data[col].fillna('Unknown')

# Add flags for imputed values
data['SpectralClass_Missing'] = (data['Spectral_Class'] == 'Unknown').astype(int)

# Save cleaned data to a new file
data.to_csv('constellations_cleaned (1).csv', index=False)
print("Cleaned data saved to constellations_cleaned (1).csv")

```

**Hypothesis:**  "Given the properties of a star, such as co-ordinates, relative distance from earth and illuminance, can we predict the constellation it belongs to?"
In my hypothesis, I am trying to assign the constellation a star belongs to, hence it is a classification task.

High-level Analysis:

Overall, through out the project I intend to build a Machine Learning model that is capable of predicting which constellation a star belongs to based on its properties. I will use key features/variables in my dataset such as co-ordinates, distance and illuminance to train a supervised classification model. The end goal is to learn the patterns that differentiate stars across constellations and evaluate how accurately the model can assign the new stars to correct constellation.

**Algorithm**


---

For the inital model, I will use the k-Nearest Neighbour classification algorithm. It is a distance-based method that assigns a class label to a point based on the majority of its k-closest neighbours in feature space.

**Assumptions of kNN**

* kNN assumes that the distance between data points reflects similarity, which aligns with astronomy since stars belonging to a constellation tend to cluster in co-ordinate space.
But in order for the Euclidean distances to be meaningful, all numerical features must be on comparable scales.
* There is no strict distribution assumed by kNN which makes it suitable for irregular, spatially clustered data.
* Although kNN does not assume that there are no outliers, it is sensitive to outliers as it heavily relies on distance. Outliers can potentially distort neighbour relationships and lead to misclassifications.
* kNN works optimally when the dataset is not extremely large and data points are dense enough to provide meaningful neighbors. The current dataset with around 11,000 samples is well within the practical size range of kNN.
* kNN does not require any independence of features and in fact tolerates correlated features far better than linear models.

**Issues with kNN**
1. kNN uses Euclidean distance, so unscaled features dominate the distance calculation. We can use StandardScaler to scale the star distances if needed.
2. Isolated points far from clusters can distort nearest neighbours, in case of our dataset this might be very distant galaxies posing as stars. In order to avoid this, either robust scaling or a different k value should be used.
3. When k is too small, overfitting may occur. This can be avoided by performing cross-validation to choose an appropriate k.

**Why kNN is a good model for this Task?**

kNN complements few features and patterns in the dataset which makes it suitable for this ML task.
* Constellations form spatial clusters in RA and Dec, which can be captured by kNN.
* It is a distance-based model which directly aligns with how constellations are defined in the night sky.
* It can handle non-linear, irregularly shaped boundaries that the constellation regions have.

**Hyperparameters for kNN**


---

kNN has three main hyperparameters to be chosen according to the application:
1. k : Number of neighbours
2. Distance Metric
3. Weight Function

**Value of k:**
* This decides the number of nearest points to be considered while classifying.
* Smaller values of k are sensitive to noise(overfitting) and larger values of k lead to underfitting.
* Inorder to properly tune it, values from 1 to 25 should be tested and the one with highest cross-validation accuracy should be chosen.

**Distance Metric**
* This decides how we measure similarity between points.
* Since most of the features in the dataset chosen for kNN are continuous, Euclidean or Manhattan would be a good fit for the distance metric.

**Weight Function**
* This decides whether all the neighbours should be given equal importance.
* Uniform is when all neighbours are given equal importance and Distance is in the case of closer neighbours getting higher importance.

These hyperparameters will be tuned by testing different combinations and selecting the one with highest validation accuracy. This can be achieved with the help of Grid Search.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd

X = df[['RA_deg', 'Dec_deg', 'Vmag', 'AbsMag', 'Distance_ly', 'SpectralClass_Missing']]
y = df['Constellation']

param_grid = {'n_neighbors': [1, 3, 5, 7, 9, 11, 15, 21],
              'metric': ['euclidean', 'manhattan'],
              'weights': ['uniform', 'distance']}

knn = KNeighborsClassifier()

grid = GridSearchCV(knn, param_grid, cv = 5, scoring = 'accuracy', n_jobs=-1)
grid.fit(X,y)

print("Best hyperparameters : ", grid.best_params_)
print("Best cross-validation accuracy : ", grid.best_score_)
```

On implementation of grid search it is evident that the most suitable Hyperparameters are as follows:

1. k value = 7
2. Distance metric : Manhattan
3. Weights : Distance

**Post-processing Techniques**


---
The dataset contains astronomical variables which have different ranges. Even after scaling before kNN, the outputs may need interpretation. Class imbalance exists as some constellations are huge with numerous stars an others are small. The following post-processing techniques are required:
1. Probability Calibration:
* KNN does not produce true probabilities, but only gives the fraction of neighbors voting for a class.
* These raw probabilities are not caliberated and can be misleading.
* Caliberation makes the output probabilities meaningful.

2. Class Imbalance Handling:
* Some constellations have many stars, while others have very few.
* kNN favors the majority classes as they appear more often in nearest neighbours.

**Final ML Model (Tuned XGBoost)**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
X = df[['RA_deg', 'Dec_deg', 'Distance_ly', 'AbsMag', 'Vmag']]
y = df['Constellation']

le = LabelEncoder()
y_enc = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size = 0.2, random_state = 42, stratify = y_enc)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

param_dist = {
    "n_estimators":    [400, 600, 800],
    "max_depth":       [4, 6, 8],
    "learning_rate":   [0.03, 0.05, 0.08],
    "subsample":       [0.7, 0.85, 1.0],
    "colsample_bytree":[0.6, 0.8, 1.0],
    "min_child_weight":[1, 3, 5],
    "gamma":           [0.0, 0.1, 0.3],
    "reg_lambda":      [1.0, 3.0, 5.0],
    "reg_alpha":       [0.0, 0.5, 1.0],
    "max_delta_step":  [0, 1, 2],

}

n_iter = 20
n_splits = 3
rng = np.random.RandomState(42)

skf = StratifiedKFold(n_splits = n_splits, shuffle = True, random_state = 42)

def sample_params(rng):
  return{
      "n_estimators":       rng.choice(param_dist["n_estimators"]),
      "max_depth":          rng.choice(param_dist["max_depth"]),
      "learning_rate":      rng.choice(param_dist["learning_rate"]),
      "subsample":          rng.choice(param_dist["subsample"]),
      "colsample_bytree":   rng.choice(param_dist["colsample_bytree"]),
      "min_child_weight":   rng.choice(param_dist["min_child_weight"]),
      "gamma":              rng.choice(param_dist["gamma"]),
      "reg_lambda":         rng.choice(param_dist["reg_lambda"]),
      "reg_alpha":          rng.choice(param_dist["reg_alpha"]),
      "max_delta_step":     rng.choice(param_dist["max_delta_step"])
  }

best_score = -np.inf
best_params = None

for i in range(1, n_iter + 1):
    params = sample_params(rng)
    fold_scores = []

    print(f"\n=== Config {i}/{n_iter} ===")
    print("Params: ", params)

    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(X_train_scaled, y_train), start = 1):
      X_tr, X_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
      y_tr, y_val = y_train[train_idx], y_train[val_idx]

      model = XGBClassifier(
          objective = 'multi:softprob',
          num_class = len(le.classes_),
          tree_method = 'hist',
          eval_metric = 'mlogloss',
          random_state = 42,
          **params
      )

      model.fit(X_tr, y_tr, verbose = False)
      y_tr_pred = model.predict(X_tr)
      y_val_pred = model.predict(X_val)
      acc_tr = accuracy_score(y_tr, y_tr_pred)
      acc = accuracy_score(y_val, y_val_pred)
      fold_scores.append(acc)

      print(f" Fold {fold_idx}/{n_splits} finished, train accuracy = {acc_tr:.4f} val accuracy = {acc: .4f}")
mean_acc = np.mean(fold_scores)
print(f"Mean CV accuracy for config {i}: {mean_acc: .4f}")
if mean_acc > best_score:
      best_score = mean_acc
      best_params = params
      print("new best config!")
print("\nBest CV accuracy: ", best_score)
print("Best params: ", best_params)

best_xgb = XGBClassifier(
      objective = 'multo:softprob',
      num_class = len(le.classes_),
      tree_method = 'hist',
      eval_metric = 'mlogloss',
      random_state = 42,
      **best_params

  )

print("\n Training final model with best params")
best_xgb.fit(X_train_scaled, y_train, verbose = False)
y_train_pred = best_xgb.predict(X_train_scaled)
train_acc = accuracy_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred, average = 'macro')
print(f"Train accuracy : {train_acc: .4f}")
print(f"Train F1 scsore: {train_f1}")

y_pred = best_xgb.predict(X_test_scaled)

acc_test = accuracy_score(y_test, y_pred)
print("Test Accuracy: ", acc_test)
f1_test = f1_score(y_test, y_pred, average = 'macro')
print(f"Test F1 score: {f1_test}")
```

In this Tuned XGBoost model, the hyperparameters were tuned with stratified cross-validation, which produced about 0.7 test accuracy. It offers the best balance between performance and overfitting.

Analysis:

* The train accuracy and train macro F1 are around 0.84, which shows that the model fits the training data well but no longer memorises it.
* The final test accuracy and macro are at 0.7 which is noticeably better than the previous models.
* The train-test gap in macro F1 indicates some expected overfitting but a much healthier bias-variance trade-off. The cross-validation in search loop helped in steering the model away from overfitting.
